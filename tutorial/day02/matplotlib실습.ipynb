```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
sr = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10)) 
```

```python
sr
```

```python
sr.plot() 
plt.savefig('figure1.svg')
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0),
                  columns=["A", "B", "C", "D"],
                  index=np.arange(0, 100, 10)) 
df
```

```python
df.plot()
plt.savefig('f2.png')
```

```python
df['A'].plot() 
```
